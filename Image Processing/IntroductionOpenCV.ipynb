```python
# cv2 is the standard library alias used for OpenCV in Python
# we get all the major image processing functions from cv2
import cv2
```

```python
img = cv2.imread("msd.jpeg")
```

```python
type(img)
```

```python
img.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
cv2.imshow("msd",img)
# Wait indefinitely until a key is pressed
# 0 means wait forever
# Without this, the window will close immediately
cv2.waitKey(0)
# close all openCV windows
# important to release memory
cv2.destroyAllWindows()
```

```python
# Convert image into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("Gray MSD",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
