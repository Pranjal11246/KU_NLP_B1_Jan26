```python
# Text -> Tokenization -> Sequences -> Embedding -> LSTM -> Prediction
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sentences = [
    "i really liked the movie",
    "the film was amazing",
    "i like the first part but not second",
    "movie was terrible",
    "i really enjoyed the movie",
    "the movie was boring",
    "i hate that film"
]
# 1 - positive, 0 - negative
labels = np.array([1, 1, 1, 0, 1, 0, 0])
```

```python
# create object of Tokenizer
tokenizer = Tokenizer()

# tokenize sentences into words
tokenizer.fit_on_texts(sentences)
```

```python
# Convert text -> numbers
sequences = tokenizer.texts_to_sequences(sentences)
```

```python
print(sequences)
```

```python
# Make all sequences same length by adding 0
X = pad_sequences(sequences)
```

```python
vocab_size = len(tokenizer.word_index) + 1
max_length = X.shape[1]
```

```python
# Sequential groups a linear stack of layers into a Model.
model = Sequential([
    # Embedding Layer
    Embedding(input_dim=vocab_size, output_dim=10, input_length=max_length),
    # LSTM Layer (sequence understanding with gates)
    LSTM(32),
    Dense(1, activation="sigmoid")
])
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, labels, epochs=100)
```

```python
def predict_sentiment(text):
  seq = tokenizer.texts_to_sequences([text])
  padded = pad_sequences(seq, maxlen=max_length)
  prediction = model.predict(padded)
  print(prediction)
  return "Positive" if prediction > 0.5 else "Negative"
```

```python
test_sentence = "movie was amazing"
prediction = predict_sentiment(test_sentence)
print(prediction)
```
